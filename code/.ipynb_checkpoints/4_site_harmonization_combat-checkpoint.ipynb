```python
# Site harmonization on gradients scores using Combat before running linear models (03-2)

import numpy as np
import pandas as pd
import os
import nibabel as nib
import matplotlib.pyplot as plt
import seaborn as sns

from neuroCombat import neuroCombat
```

```python
# Load gradient scores of individuals

root_pth = '/Users/valkebets/Documents/Work/project-data/project__sensory_gradients/analyses_clean'
results_pth = root_pth + '/results/gradients_ctx_thal_ipsiLR_sym-thal_aligned_abide1_N211/files'

#data = np.load(results_pth + '/ctx_ind_loadings/ind_g1-g2_ctx_thal_loadings.npz')
#all_g1_ind_cortex = data['all_g1_ind_ctx_thal_lhrh'] 
#all_g1_ind_thal = data['all_g1_ind_thal_ctx_lhrh']
#all_g2_ind_cortex = data['all_g2_ind_ctx_thal_lhrh']
#all_g2_ind_thal = data['all_g2_ind_thal_ctx_lhrh']

data = np.load(results_pth + '/thal_ind_loadings/ind_g1-g2_norm-angle_notaligned_thal_loadings.npz')
all_g1_scores = data['all_g1_ind_thal_lhrh']
all_g2_scores = data['all_g2_ind_thal_lhrh']

print('Dimensions of G1 thalamic scores:', all_g1_scores.shape) # should be 211 x 648
print('Dimensions of G2 thalamic scores:', all_g2_scores.shape) # should be 211 x 648

#print('Dimensions of G1 cortical scores:', all_g1_ind_cortex.shape) # should be 211 x 400
#print('Dimensions of G1 thalamic scores:', all_g1_ind_thal.shape) # should be 211 x 648

#print('Dimensions of G2 cortical scores:', all_g2_ind_cortex.shape) # should be 211 x 400
#print('Dimensions of G2 thalamic scores:', all_g2_ind_thal.shape) # should be 211 x 648

# Concatenate together cortical + thalamic gradient scores >> no need, not using cortical scores
#all_g1_scores = np.concatenate([all_g1_ind_cortex, all_g1_ind_thal], axis=1)
#all_g2_scores = np.concatenate([all_g2_ind_cortex, all_g2_ind_thal], axis=1)

#print('Dimensions of G1 scores (cortex+thalamus):', all_g1_scores.shape) # should be 211 x 1048 (400+648)
#print('Dimensions of G2 scores (cortex+thalamus):', all_g2_scores.shape) # should be 211 x 1048 (400+648)
```

```python
# Check min/max gradient scores

print('Min/Max G1 thalamic scores:', np.min(all_g1_scores), np.max(all_g1_scores))
print('Min/Max G2 thalamic scores:', np.min(all_g2_scores), np.max(all_g2_scores))

#print('Min/Max G1 cortical scores:', np.min(all_g1_ind_cortex), np.max(all_g1_ind_cortex))
#print('Min/Max G1 thalamic scores:', np.min(all_g1_ind_thal), np.max(all_g1_ind_thal))

#print('Min/Max G2 cortical scores:', np.min(all_g2_ind_cortex), np.max(all_g2_ind_cortex))
#print('Min/Max G2 thalamic scores:', np.min(all_g2_ind_thal), np.max(all_g2_ind_thal))
```

```python
# Load demographics with subject IDs

df =  pd.read_csv(root_pth + '/data/abide_func_fn_abideI.csv', header=0) #index_col=0)

# Keep only selected columns
df = df[['ID','OrgID','Group','Age','Site','Func_MeanFD']]

# Remove subject with FD>0.3
df = df.loc[(df['Func_MeanFD'] < 0.3)]

# Move subject 32183 to end
subj_idx = df.index[df.ID == 32183]
subj_idx_list = subj_idx.tolist()
#print('Index of subject 32183:', subj_idx_list)

def shift_row_to_bottom(df, index_to_shift):
    idx = [i for i in df.index if i!=index_to_shift]
    return df.loc[idx+[index_to_shift]]

df = shift_row_to_bottom(df, 172)

# Reset index
df.reset_index(drop=True, inplace=True)

# Convert ID & OrigID back to integers
df.ID = df.ID.astype(int)
df.OrgID = df.OrgID.astype(int)

# Get group indices
asd_idx = df.index[df['Group'] == 'ASD'].tolist()
print(len(asd_idx), 'ASD subjects')

td_idx = df.index[df['Group'] == 'CONTROL'].tolist()
print(len(td_idx), 'TD subjects')

df
```

```python
# Find boundaries between sites (for plotting it later)

print('Sites :', np.unique(df.Site))

nyu_idx = df.index[df['Site'] == 'NYU'].tolist()
pitt_idx = df.index[df['Site'] == 'Pitt'].tolist()
usm_idx = df.index[df['Site'] == 'USM'].tolist()

print('Pitt index :', pitt_idx[0], '-', pitt_idx[-1], '(n =', len(pitt_idx),')')
print('USM index :', usm_idx[0], '-', usm_idx[-1], '(n =', len(usm_idx),')')
print('NYU index :', nyu_idx[0], '-', nyu_idx[-1], '(n =', len(nyu_idx),')')
```

```python
# Recode group & site to numerical values

df['Group'] = pd.Categorical(df.Group).codes # ASD=0, TD=1
df['Site'] = pd.Categorical(df.Site).codes # NYU=0, Pitt=1, USM=2 (recodes by alphabetical order)
#df = df.replace("ASD", 1) # alternative way of turning categorical into numerical values

print('Group :', np.unique(df.Group))
print('Sites :', np.unique(df.Site))

# Check if number of subjects is correct after numerisation of site info
nyu_idx = df.index[df['Site'] == 0].tolist()
print('NYU n =', len(nyu_idx))

pitt_idx = df.index[df['Site'] == 1].tolist()
print('Pitt n =', len(pitt_idx))

usm_idx = df.index[df['Site'] == 2].tolist()
print('USM n =', len(usm_idx))

df
```

```python
# Combat harmonization: remove effect of site while preserving effect of age
# G1

# https://github.com/Jfortin1/neuroCombat/tree/ac82a067412078680973ddf72bd634d51deae735

data1 = all_g1_scores.T
print('Dimensions of data:', data1.shape)

# Specifying the batch (scanner variable) as well as a biological covariate to preserve:
covars = df[['Site','Group','Age']]
print('Dimensions of site & group:', covars.shape)

# To specify names of the variables that are categorical:
categorical_cols = ['Group']
continuous_cols = ['Age']

# To specify the name of the variable that encodes for the scanner/batch covariate:
batch_col = 'Site'

#Harmonization step:
data1_combat = neuroCombat(dat=data1,
    covars=covars,
    batch_col=batch_col,
    categorical_cols=categorical_cols,
    continuous_cols=continuous_cols)["data"]

data1_combat
```

```python
# Plot distribution of gradient scores before/after ComBat

print(np.min(data1.flatten()), np.max(data1.flatten()))
print(np.min(data1_combat.flatten()), np.max(data1_combat.flatten()))

plt.figure()
plt.subplot(121)
plt.hist(data1.flatten(), bins=40, color='midnightblue', range=(-0.3, 0.3))
plt.title('Before ComBat')
#plt.show()

plt.subplot(122)
plt.hist(data1_combat.flatten(), bins=40, color='midnightblue', range=(-0.3, 0.3))
plt.title('After ComBat')
plt.show()
```

```python
# Plot G1 scores before & after harmonization

print('Site index boundaries at subjects:', pitt_idx[-1], usm_idx[-1])

plt.figure()

plt.subplot(121)
heat_map = sns.heatmap(data1, vmin=-0.2, vmax=0.2)
plt.axvline(x=38, color='white') # vertical line at boundaries between sites
plt.axvline(x=124, color='white') # vertical line at boundaries between sites
plt.xticks([])
plt.yticks([])
plt.xlabel('Subjects')
plt.ylabel('Gradient scores')
plt.title('Before ComBat')

plt.subplot(122)
heat_map2 = sns.heatmap(data1_combat, vmin=-0.2, vmax=0.2)
plt.axvline(x=38, color='white') # vertical line at boundaries between sites
plt.axvline(x=124, color='white') # vertical line at boundaries between sites
plt.xticks([])
plt.yticks([])
plt.xlabel('Subjects')
#plt.ylabel('Gradient scores')
plt.title('After ComBat')
plt.show()
```

```python
# Combat harmonization: remove effect of site while preserving effect of age
# G2

data2 = all_g2_scores.T
print('Dimensions of data:', data2.shape)

# Specifying the batch (scanner variable) as well as a biological covariate to preserve:
covars = df[['Site','Group','Age']]
print('Dimensions of site & group:', covars.shape)

# To specify names of the variables that are categorical:
categorical_cols = ['Group']
continuous_cols = ['Age']

# To specify the name of the variable that encodes for the scanner/batch covariate:
batch_col = 'Site'

#Harmonization step:
data2_combat = neuroCombat(dat=data2,
    covars=covars,
    batch_col=batch_col,
    categorical_cols=categorical_cols,
    continuous_cols=continuous_cols)["data"]

data2_combat
```

```python
# Plot G2 scores before & after harmonization

print('Site index boundaries at subjects:', pitt_idx[-1], usm_idx[-1])

plt.figure()

plt.subplot(121)
heat_map = sns.heatmap(data2, vmin=-0.2, vmax=0.2)
plt.axvline(x=38, color='white') # vertical line at boundaries between sites
plt.axvline(x=124, color='white') # vertical line at boundaries between sites
plt.xticks([])
plt.yticks([])
plt.xlabel('Subjects')
plt.ylabel('Gradient scores')
plt.title('Before combat')

plt.subplot(122)
heat_map2 = sns.heatmap(data2_combat, vmin=-0.2, vmax=0.2)
plt.axvline(x=38, color='white') # vertical line at boundaries between sites
plt.axvline(x=124, color='white') # vertical line at boundaries between sites
plt.xticks([])
plt.yticks([])
plt.xlabel('Subjects')
#plt.ylabel('Gradient scores')
plt.title('After combat')
plt.show()
```

```python
all_g1_ind_thal_lhrh_combat = data1_combat.T
all_g2_ind_thal_lhrh_combat = data2_combat.T

print(all_g1_ind_thal_lhrh_combat.shape)
print(all_g2_ind_thal_lhrh_combat.shape)
```

```python
# Get gradients for subjects from each group

all_g1_ind_thal_lhrh_combat_asd = all_g1_ind_thal_lhrh_combat[asd_idx,:]
all_g1_ind_thal_lhrh_combat_td = all_g1_ind_thal_lhrh_combat[td_idx,:]
all_g2_ind_thal_lhrh_combat_asd = all_g2_ind_thal_lhrh_combat[asd_idx,:]
all_g2_ind_thal_lhrh_combat_td = all_g2_ind_thal_lhrh_combat[td_idx,:]

print('Dimensions of individual G1 thalamic scores for ASD subjects:', all_g1_ind_thal_lhrh_combat_asd.shape) 
print('Dimensions of individual G2 thalamic scores for ASD subjects:', all_g2_ind_thal_lhrh_combat_asd.shape) 

print('Dimensions of individual G1 thalamic scores for TD subjects:', all_g1_ind_thal_lhrh_combat_td.shape) 
print('Dimensions of individual G2 thalamic scores for TD subjects:', all_g2_ind_thal_lhrh_combat_td.shape)
```

```python
# Average individual gradient scores for each group

mean_g1_ind_thal_lhrh_combat_asd = all_g1_ind_thal_lhrh_combat_asd.mean(0)
mean_g1_ind_thal_lhrh_combat_td = all_g1_ind_thal_lhrh_combat_td.mean(0)

print('Dimensions of group-averaged gradient:') # should be 648
print('ASD: ', mean_g1_ind_thal_lhrh_combat_asd.shape)
print('TD: ', mean_g1_ind_thal_lhrh_combat_td.shape)

# Same for G2
mean_g2_ind_thal_lhrh_combat_asd = all_g2_ind_thal_lhrh_combat_asd.mean(0)
mean_g2_ind_thal_lhrh_combat_td = all_g2_ind_thal_lhrh_combat_td.mean(0)
```

```python
# ASD group loadings 
# not trying with new mask anymore since same results

atlas_nii = nib.load(root_pth + '/data/thalamus_intersectLR_mask.nii')
atlas = np.asanyarray(atlas_nii.dataobj)
m = np.isin(atlas, [10, 49])

# Save G1 scores as nifti
g1_ref_thal = np.zeros_like(atlas)
g1_ref_thal[m] = mean_g1_ind_thal_lhrh_combat_asd
g1_ref_thal_nii = nib.Nifti1Image(g1_ref_thal, atlas_nii.affine, header=atlas_nii.header)
nib.save(g1_ref_thal_nii, results_pth + '/thal_group_loadings/g1_norm-angle_notaligned_thal_mean_loadings_combat_asd.nii')

# Save G2 scores as nifti
g2_ref_thal = np.zeros_like(atlas)
g2_ref_thal[m] = mean_g2_ind_thal_lhrh_combat_asd
g2_ref_thal_nii = nib.Nifti1Image(g2_ref_thal, atlas_nii.affine, header=atlas_nii.header)
nib.save(g2_ref_thal_nii, results_pth + '/thal_group_loadings/g2_norm-angle_notaligned_thal_mean_loadings_combat_asd.nii')
```

```python
# TD group loadings

# Save G1 scores as nifti
g1_ref_thal = np.zeros_like(atlas)
g1_ref_thal[m] = mean_g1_ind_thal_lhrh_combat_td
g1_ref_thal_nii = nib.Nifti1Image(g1_ref_thal, atlas_nii.affine, header=atlas_nii.header)
nib.save(g1_ref_thal_nii, results_pth + '/thal_group_loadings/g1_norm-angle_notaligned_thal_mean_loadings_combat_td.nii')

# Save G2 scores as nifti
g2_ref_thal = np.zeros_like(atlas)
g2_ref_thal[m] = mean_g2_ind_thal_lhrh_combat_td
g2_ref_thal_nii = nib.Nifti1Image(g2_ref_thal, atlas_nii.affine, header=atlas_nii.header)
nib.save(g2_ref_thal_nii, results_pth + '/thal_group_loadings/g2_norm-angle_notaligned_thal_mean_loadings_combat_td.nii')
```
