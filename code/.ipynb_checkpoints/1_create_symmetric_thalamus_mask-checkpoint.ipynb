## Create a symmetric thalamus mask


First, we create a symmetric mask of the thalamus, by taking the intersection between the left and right thalamus masks, and flipping

```python
# Create symmetric intersection between left & right thalamus
# try creating copy of the flipped mask

import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
data_dir = '/Users/valkebets/Documents/Work/project-data/project__sensory_gradients/analyses_clean/data/'
img = nib.load(data_dir + 'Atlas_ROIs.3.nii.gz')
subcortex = img.get_fdata()
```

```python
# Plot subcortical atlas

a = plt.imshow(subcortex[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(a)
plt.title('Subcortical atlas')
plt.show()


# Try flipping subcortical atlas

from numpy import flipud

subcortex_flipped = flipud(subcortex)

b = plt.imshow(subcortex_flipped[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(b)
plt.title('Flipped subcortical atlas')
plt.show()
```

```python
# Select only thalamus 

thal_bin = np.where(((subcortex==10) | (subcortex==49)), 1, 0)      
thal_flipped_bin = np.where(((subcortex_flipped==10) | (subcortex_flipped==49)), 1, 0)

print('Dimensions thalamus mask =', thal_bin.shape)
print('Dimensions flipped thalamus mask =',thal_flipped_bin.shape)
```

```python
# Plot left + right thalamus
c = plt.imshow(thal_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(c)
plt.title('Bilateral thalamus mask')
plt.show()

# Plot flipped thalamus mask
d = plt.imshow(thal_flipped_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(d)
plt.title('Flipped bilateral thalamus mask')
plt.show()
```

```python
# Plot only left thalamus

thal_lh_bin = np.where((subcortex==10), 1, 0)
                     
e = plt.imshow(thal_lh_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(e)
plt.title('Left thalamus mask')
plt.show()


# Flip left thalamus mask >>> making a copy
#thal_lh_flipped_bin = flipud(thal_lh_bin)
thal_lh_flipped_bin = flipud(thal_lh_bin).copy()


f = plt.imshow(thal_lh_flipped_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(f)
plt.title('Flipped left thalamus mask')
plt.show()
```

```python
# Plot right thalamus

thal_rh_bin = np.where((subcortex==49), 1, 0)

g = plt.imshow(thal_rh_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(g)
plt.title('Right thalamus mask')
plt.show()

# Flip right thalamus mask >>> making a copy

#thal_rh_flipped_bin = flipud(thal_rh_bin)
thal_rh_flipped_bin = flipud(thal_rh_bin)

i = plt.imshow(thal_rh_flipped_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(i)
plt.title('Flipped right thalamus mask')
plt.show()
```

```python
# Find intersection between left thalamus & flipped right thalamus
# Find intersection between right thalamus & flipped left thalamus

print('Dimensions left thalamus mask =', thal_lh_bin.shape, np.sum(thal_lh_bin), 'voxels')
print('Dimensions right flipped thalamus mask =', thal_rh_flipped_bin.shape, np.sum(thal_rh_flipped_bin), 'voxels')
common_thal_lh = np.where(((thal_lh_bin==1) & (thal_rh_flipped_bin==1)), 10, 0)  # used for later
common_thal_lh_bin = np.where(((thal_lh_bin==1) & (thal_rh_flipped_bin==1)), 1, 0)  # just to check voxels
print('Dimensions common left thalamus mask =', common_thal_lh.shape, np.sum(common_thal_lh_bin), 'voxels')
print('')

print('Dimensions right thalamus mask =', thal_rh_bin.shape, np.sum(thal_rh_bin), 'voxels')
print('Dimensions left flipped thalamus mask =', thal_lh_flipped_bin.shape, np.sum(thal_lh_flipped_bin), 'voxels')
common_thal_rh = np.where(((thal_rh_bin==1) & (thal_lh_flipped_bin==1)), 49, 0)  # used for later 
common_thal_rh_bin = np.where(((thal_rh_bin==1) & (thal_lh_flipped_bin==1)), 1, 0)  # just to check voxels
print('Dimensions common right thalamus mask =', common_thal_rh.shape, np.sum(common_thal_rh_bin), 'voxels')
```

```python
# Plot intersection between left thalamus & flipped right thalamus

j = plt.imshow(common_thal_lh[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(j)
plt.title('Common left thalamus mask')
plt.show()

k = plt.imshow(common_thal_rh[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(k)
plt.title('Common right thalamus mask')
plt.show()
```

```python
# Join the two common thalamus masks

common_thal_lhrh = np.add(common_thal_lh, common_thal_rh)

print('Dimensions common bilateral thalamus mask =', common_thal_lhrh.shape)

l = plt.imshow(common_thal_lhrh[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(l)
plt.title('Common bilateral thalamus mask')
plt.show()
```

```python
np.shares_memory(common_thal_lh, common_thal_rh)
```

```python
print(np.unique(common_thal_lhrh))
idx_common_lhrh = np.array(np.where(common_thal_lhrh!=0))
#print(idx_common_lhrh)
#print(idx_common_lhrh.shape)
```

```python
# Binarize mask

common_thal_lhrh_bin = np.zeros_like(common_thal_lhrh)
common_thal_lhrh_bin[common_thal_lhrh!=0] = 1
#= np.where(((common_thal_lhrh==10) & (common_thal_lhrh==49)), 1, 0)  
print(np.unique(common_thal_lhrh_bin))

idx_common_lhrh_bin = np.array(np.where(common_thal_lhrh_bin!=0))
#print(idx_common_lhrh)
print(idx_common_lhrh_bin.shape)
```

```python
l = plt.imshow(common_thal_lhrh_bin[10:50, 39, 10:40].T,
               cmap='viridis',
               origin='lower',
              )
plt.axis('off')
plt.colorbar(l)
plt.title('Common bilateral thalamus mask')
plt.show()
```

```python
# Save the binary mask

atlas = np.asanyarray(img.dataobj)
m = np.isin(atlas, [10, 49])
new_mask = np.zeros_like(atlas)
new_mask = nib.Nifti1Image(common_thal_lhrh_bin, img.affine, header=img.header)
nib.save(new_mask, data_dir + '/thalamus_intersectLR_mask_new2.nii')
```
