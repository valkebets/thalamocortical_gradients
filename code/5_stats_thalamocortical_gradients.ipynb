## Compute linear models assessing group differences in thalamocortical gradients

* Using gradient scores after site harmonization (ComBat)
* Models are adjusting for age and head motion
* Removed the code for cortico-thalamic gradients in cortex space

```python
import numpy as np
import pandas as pd
import os
import nibabel as nib
import matplotlib.pyplot as plt

from brainstat.stats.terms import FixedEffect, MixedEffect
from brainstat.stats.SLM import SLM
```

```python
# Load individual gradient scores

root_pth = '/Users/valkebets/Documents/Work/project-data/project__sensory_gradients/analyses_clean'
results_pth = root_pth + '/results/gradients_ctx_thal_ipsiLR_sym-thal_aligned_abide1_N211/files'

data = np.load(results_pth + '/thal_ind_loadings/ind_g1-g2_norm-angle_notaligned_thal_loadings_combat.npz')
all_grad_scores = data['data2_combat'].T
print('Dimensions of thalamic gradient scores:', all_grad_scores.shape) # should be 211 x 648
```

```python
# Load demographics with subject IDs
df =  pd.read_csv(root_pth + '/data/abide_func_fn_abideI.csv', header=0) #index_col=0)

# Keep only selected columns
df = df[['ID','OrgID','Group','Age','Site','Func_MeanFD']]

# Remove subject with FD>0.3
df = df.loc[(df['Func_MeanFD'] < 0.3)]

# Move subject 32183 to end
subj_idx = df.index[df.ID == 32183]
subj_idx_list = subj_idx.tolist()
#print('Index of subject 32183:', subj_idx_list)

def shift_row_to_bottom(df, index_to_shift):
    idx = [i for i in df.index if i!=index_to_shift]
    return df.loc[idx+[index_to_shift]]

df = shift_row_to_bottom(df, 172)

# Reset index
df.reset_index(drop=True, inplace=True)

# Convert ID & OrigID back to integers
df.ID = df.ID.astype(int)
df.OrgID = df.OrgID.astype(int)

# Get group indices
asd_idx = df.index[df['Group'] == 'ASD'].tolist()
print(len(asd_idx), 'ASD subjects')

td_idx = df.index[df['Group'] == 'CONTROL'].tolist()
print(len(td_idx), 'TD subjects')

df
```

```python
# Test with simple linear model for age and group

term_age = FixedEffect(df.Age)
term_patient = FixedEffect(df.Group)
model = term_age + term_patient

print(model)
```

```python
# Group differences using linear model with group and age

term_age = FixedEffect(df.Age)
term_patient = FixedEffect(df.Group)

# treating head motion as fixed effect
term_motion = FixedEffect(df.Func_MeanFD)

#model = term_age + term_patient
model = term_age + term_patient + term_motion

contrast_patient = model.Group_CONTROL - model.Group_ASD

slm_linear = SLM(
    model,
    contrast_patient,
    correction="fdr")
    #two_tailed=False)

slm_linear.fit(all_grad_scores)

print(slm_linear)
```

```python
# Group differences using mixed effects linear model while correcting for age (fixed effect) 
# and head motion, along with FDR correction


term_age = FixedEffect(df.Age)
term_patient = FixedEffect(df.Group)
random_motion = MixedEffect(df.Func_MeanFD, name_ran="Func_MeanFD") 

model_random = term_age + term_patient + random_motion
contrast_patient = model.Group_CONTROL - model.Group_ASD

slm_random = SLM(
    model_random,
    contrast_patient,
    correction="fdr")

slm_random.fit(all_grad_scores)
print(slm_random)
```

```python
# Check t-vals

tvals = slm_random.t
#tvals = tvals[0,:]
tvals = tvals.flatten()
print('Dimensions of t-vals:', tvals.shape)
```

```python
# Plot t-values on histogram

plt.hist(tvals, bins=40, color='midnightblue')
plt.show()
```

```python
# Check FDR-corrected q vals

qvals = slm_random.Q
print('Dimensions of q-vals:', qvals.shape)

# Check if anything survived to FDR test
print('Min p-value after FDR correction = ', min(qvals))

# remove p-vals > 0.05
qvals[qvals > 0.05] = np.nan
```

```python
# # Plot q-values on histogram

plt.hist(qvals, bins=10, color='midnightblue', range=(0,0.05))
plt.show()
```

```python
# Save thalamic t-vals to nifti

atlas_nii = nib.load(root_pth + '/data/thalamus_intersectLR_mask.nii')
atlas = np.asanyarray(atlas_nii.dataobj)
m = np.isin(atlas, [10, 49]) # thalamus mask
grad_t = np.zeros_like(atlas)
grad_t[m] = tvals
grad_t_nii = nib.Nifti1Image(grad_t, atlas_nii.affine, header=atlas_nii.header)
nib.save(grad_t_nii, results_pth + '/thal_stats_mixed-effects/g2_thal_norm-angle_notaligned_tvals_combat.nii')
```

```python
# Save thalamic q-vals to nifti

atlas_nii = nib.load(root_pth + '/data/thalamus_intersectLR_mask.nii')
atlas = np.asanyarray(atlas_nii.dataobj)
m = np.isin(atlas, [10, 49]) # thalamus mask
grad_q = np.zeros_like(atlas)
grad_q[m] = qvals
grad_q_nii = nib.Nifti1Image(grad_q, atlas_nii.affine, header=atlas_nii.header)
nib.save(grad_q_nii, results_pth + '/thal_stats_mixed-effects/g2_thal_norm-angle_notaligned_qvals_combat.nii')
```
