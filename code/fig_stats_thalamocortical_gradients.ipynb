```python
# Plot t-vals & q-vals on volume (thalamic FC group differences)

import numpy as np
import pandas as pd
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
root_pth = '/Users/valkebets/Documents/Work/project-data/project__sensory_gradients/analyses_clean'
results_pth = root_pth + '/results/gradients_ctx_thal_ipsiLR_sym-thal_aligned_abide1_N211/files'

tvals = nib.load(results_pth + '/thal_stats_mixed-effects/rg2_thal_norm-angle_notaligned_tvals_combat.nii')
grad_img_data = tvals.get_fdata()
print('Dimensions of thalamus loadings: ', grad_img_data.shape)

# Load binary mask to only consider gradient values that are not zero
mask_img = nib.load(results_pth + '/thal_stats_mixed-effects/rg2_thal_norm-angle_notaligned_tvals_combat_mask_bin.nii')
mask_img_data = mask_img.get_fdata()
print('Dimensions of mask: ', mask_img_data.shape)
print('Values in mask: ', np.unique(mask_img_data))

templ_img = nib.load(root_pth + '/data/ch2.nii')
templ_img_data = templ_img.get_fdata()
print('Dimensions of background template: ', templ_img.shape)
```

```python
#Â Mask tvals with binarized mask

grad_img_data[mask_img_data != 1] = np.nan # white background!
```

```python
# Axial views
# Load background image with gray colormap

z = 78 # 70-90
d = plt.imshow(templ_img_data[50:132, 70:140, z].T, # change to see all brain
              cmap='gray',
              origin='lower',
               alpha = 0.8)
e = plt.imshow(grad_img_data[50:132, 70:140, z].T,
               cmap='RdBu_r',
               origin='lower',
               vmin=-3,
               vmax=3)
plt.axis('off')
#plt.colorbar(e)
#plt.savefig(root_pth + '/figs/g1_tvals_ax' + str(z) + '.png')
plt.show()
```
